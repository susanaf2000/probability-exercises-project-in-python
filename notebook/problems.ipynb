## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
# TODO
import random
import numpy as np 

#create a function to roll 1000 times
def experiment(num_rolls =1000):
    greater_or_even = 0
    #create a loop where x time sums the two dices  
    for i in range(num_rolls):   
        dice1 = random.randint(1,6)
        dice2 = random.randint(1,6)
        dices = dice1 + dice2

        #after the sum, stores the value if when greater than 7 or even number
        if dices > 7 or dices % 2 == 0:
            greater_or_even += 1
    #returns by dividing it with the total of trials
    return greater_or_even / num_rolls


print(f"The probability is: {np.round(experiment() * 100, 2)}%")
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

#print(ball_box)
            
# TODO

#create a function to take the balls for each trial
def take_balls(num_trials =1000):
    #create the variables to store it after
    take1 = 0
    take2 = 0

    #for each item in the range 0,1000
    for i in range(num_trials):
        #create a list to store values after
        colors = []

        #for each item in a range of 5, because we can only take 5 balls before putting it back in the box
        for i in range(5):
            #we add the random values from ball_box keys, corresponding to the keys from 0 until 59 
            colors.append(ball_box[np.random.randint(0,59)])

        #convert the colors list to a numpy list for better filtering
        colors = np.array(colors)

        #we filter it to better usage 
        white_colors = sum(colors == "White")
        red_colors = sum(colors == "Red")
        green_colors = sum(colors == "Green")

        #use the condition we want to then store it to our take1,take2 that correspond to each step of the problem
        if white_colors == 3 and red_colors == 2:
            take1 += 1
        if white_colors == 5 or red_colors == 5 or green_colors == 5:
            take2 += 1
    
    #finally we calculate the probability
    return take1 / num_trials, take2 / num_trials

print(f"The probability of taking 3 white balls and then 2 red balls are: {np.round(take_balls()[0] * 100, 2)}%")
print(f"The probability of taking all the balls the same color is: {np.round(take_balls()[1] * 100, 2)}%")
```
